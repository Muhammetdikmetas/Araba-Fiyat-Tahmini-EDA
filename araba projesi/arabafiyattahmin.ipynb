Şimdi kullanacağımız modellerin kütüphanelerini ekleyelim Sonrada modelleri tek tek deniyecem hangi model yüksek oran verse onu seçicem 

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
data.columns
```

Önce X ve Y değerlerimizi belirtelim 

```python
y=data.arabanin_fiyati.values.reshape(-1,1)
x = data.drop(["arabanin_fiyati"], axis=1)  # reshape yapma!

x = x.select_dtypes(include=[np.number])  

```

Verileri aynı ölçeğe getirerek, farklı büyüklükteki özelliklerin modellemeyi etkilememesi için X değişkenime normalizasyon uyguladım.

```python
x = (x - x.min()) / (x.max() - x.min()) 
```

Verimi train ve test olarak ayıralım ; %80’ini eğitmek için, %20’sini ise test için kullandım. 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

Şimdi birden çok model deneyip hangisinin en yüksek sonucu verdiğine bakalım 

1.Multiple Linear Regression 

```python
model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)


print("R2 Score:", r2_score(y_test, y_pred))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
```

2.Decision Tree

```python


model_dt = DecisionTreeRegressor(random_state=42)
model_dt.fit(x_train, y_train)

y_pred_dt = model_dt.predict(x_test)

print("Decision Tree R2 Score:", r2_score(y_test, y_pred_dt))
print("Decision Tree MSE:", mean_squared_error(y_test, y_pred_dt))
```

3.Random Forest 

```python

model_rf = RandomForestRegressor(random_state=42)
model_rf.fit(x_train, y_train)

y_pred_rf = model_rf.predict(x_test)

print("Random Forest R2 Score:", r2_score(y_test, y_pred_rf))
print("Random Forest MSE:", mean_squared_error(y_test, y_pred_rf))
```

4.Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly = PolynomialFeatures(degree=2)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)

model_poly = LinearRegression()
model_poly.fit(x_train_poly, y_train)

y_pred_poly = model_poly.predict(x_test_poly)

print("Polynomial R2 Score:", r2_score(y_test, y_pred_poly))
print("Polynomial MSE:", mean_squared_error(y_test, y_pred_poly))
```

Modellere genel bakış 

```python
print("R2 Karşılaştırması:")
print("Linear Regression:", r2_score(y_test, model.predict(x_test)))
print("Decision Tree:", r2_score(y_test, y_pred_dt))
print("Random Forest:", r2_score(y_test, y_pred_rf))
print("Polynomial Regression:", r2_score(y_test, y_pred_poly))

print("\nMSE Karşılaştırması:")
print("Linear Regression:", mean_squared_error(y_test, model.predict(x_test)))
print("Decision Tree:", mean_squared_error(y_test, y_pred_dt))
print("Random Forest:", mean_squared_error(y_test, y_pred_rf))
print("Polynomial Regression:", mean_squared_error(y_test, y_pred_poly))
```
